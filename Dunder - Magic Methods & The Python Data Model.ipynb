```python
import inspect   #imported inspect library.

x = [1, 2, 3]   #x list has 3 items.
y = [4, 5]     #y list has 2 items

print(x+y) #printed aggregation of x and y list.
#print(x * 3) #x * 3 is not x items multiply 3. It is x items repeating 3 times. If you wanna see delete the number sign. 

class Person:   #created class name by "Person".
    def __init__(self, name):  #the init function is attribute of our Person class. init function need 1 parameter as "name"
        self.name = name
    def __repr__(self):     #the repr function will return our name attribute of Human class.
        return f"person({self.name})"

    def __mul__(self, x):  #the mulp function will assign x times name to name.
        if type(x) is not int:    #if you didn't write integer variable this error will showed.
            raise Exception("Invalid argument, must be int")
        self.name = self.name * x

    def __call__(self, y):     #the call function will print variable that you write 
        print(y)

    def __len__(self):
        return len(self.name)
p = Person("Mert")
p * 4     #this mean multiply function. Normally it won't work but in python language, it have to work :) 
print(p)   #repr function
p(4)  #call function
print(len(p)) #lenght of our p*4



```

```python

from queue import Queue as q #assign our FIFO Queue to q character.
class Queue(q):   #Created Queue class. It need one parameter. 
    def __repr__(self):  #repr function will return how many elements in q.
        return f"Queue({self._qsize()})"

    def __add__(self, item):  #with this add function we can put element to queue
        self.put(item)
    
    def __sub__(self, item): #with this sub function we can work with negative items
        self.get()

qu = Queue()
print(qu) #here qu hasn't any element.

qu + 7 #this item "7" will be inside of qu. 
print(qu)  #So our qu has 1 element in here.
qu + 9
qu.get(None) #if the sub function doesn't exist. This wouldn't work. So we can extract element.
print(qu) #here qu has 2 elements
#dir(qu)  #it will return valid attributes of the object.
```
