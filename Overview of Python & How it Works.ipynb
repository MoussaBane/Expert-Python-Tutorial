```python
def make_class(x): #Created the function as make_class. This function need 1 parameter as "x"
    class Human:    #Existed class inside of function.
        def __init__(self, name): #the init function is attribute of our Human class. init function need 1 parameter as "name"
            self.name = name       # self.name will return name of class


        def print_value(self):   #print_value function will print our x parameter.
            print(x)                 

    return Human

mc = make_class(10) #in this condition our parameter will be 10.
m = mc("Mert")  #our name attribute will be "Mert"
print(m.name)   #it will print our attribute.
m.print_value() #finally it will print our x parameter.
```

```python
for i in range(10): # after this sentence for loop will exist and "i" will inside of our loop.
    def show():     #we defined "show" function. This function will print "i multiply to 2".
        print(i*2)  

    show()        #if we call "show function" in this block. function will be inside of loop and will repeat 10 times and show to us every "i multiply 2"
show()   #if we call "show function" in this block. function will show mjust last number to us.
```
